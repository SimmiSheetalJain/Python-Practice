```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 2 * np.pi, 100)  # Create an array of values from 0 to 2*pi
y = np.sin(x)  # Calculate the sine values for each x

# Plot the sine wave
plt.plot(x, y, label='Sine Wave')

# Add labels and a legend
plt.title('Sine Wave Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Display the plot
plt.show()
```
