```python
def bitwise_xor_subarrays(N, A):
    result = 0

    for i in range(N):
        print(f"Iteration: {i+1}, and i = {i}, N = {N}")
        # Count subarrays in which A[i] contributes
        subarrays_count = (i + 1) * (N - i)
        print(subarrays_count)

        # Update result by XORing with A[i] * subarrays_count
        result ^= A[i] * (subarrays_count%2)
        print(result)
        print("----------------")
    return result

# Example usage
N = 4
A = [2,4,7,6]
output = bitwise_xor_subarrays(N, A)
print(output)
```
