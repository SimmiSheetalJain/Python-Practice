```python
Life_partners = "Maa loves Sheetal So Much and Sheetal Also Loves Her too Much"
```

```python
x=Life_partners.split()
print(x)
```

We can also store splited objects into variables

```python
Anjali = "Happy birthday dear"
word1, word2, word3=Anjali.split()
```

Note : The number of variable and the number of values generated by the split function should be same

```python
x=input("Enter two numbers by separating them using one charater space: ")
n1,n2 = x.split()
n1=int(n1)
n2=int(n2)
print(n1+n2)

```

#### We can pass a separator in split function By default split function has space as separater

```python
Name="Simmi#Sheetal#jain"
Name.split('#')
```
